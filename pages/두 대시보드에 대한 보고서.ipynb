# 이탈 예측 대시보드 통합 설계 보고서

## 1. 개요

본 문서는 현재 개발된 **두 개의 모델 모니터링 페이지를 분석하여 하나의 통합 페이지로 재구성하기 위한 설계 기준을 정리한 문서**이다.

현재 시스템에는 다음과 같은 두 종류의 페이지가 존재한다.

1. **모델 성능 분석 페이지**
2. **운영 모니터링 페이지**

두 페이지는 각각 다음과 같은 목적을 가진다.

| 페이지         | 목적              |
| ----------- | --------------- |
| 모델 성능 페이지   | 머신러닝 모델의 성능 평가  |
| 운영 모니터링 페이지 | 실제 서비스 운영 상태 확인 |

두 페이지를 하나로 통합함으로써 다음과 같은 목표를 달성하고자 한다.

* 모델 성능과 운영 상태를 **한 화면에서 확인**
* **데이터 사이언스 관점과 비즈니스 관점의 정보 통합**
* 중복 기능 제거
* 사용자 경험(UX) 개선

---

# 2. 두 페이지의 장점

각 페이지가 가지고 있는 장점을 분석하여 통합 페이지에 반영할 기능을 선정한다.

---

## 2.1 모델 성능 분석 페이지의 장점

모델 성능 분석 페이지는 **머신러닝 모델 평가 중심 구조**를 가지고 있다.

### 1) 직관적인 성능 지표 표시

다음과 같은 핵심 평가 지표를 카드 형태로 제공한다.

* Accuracy
* F1 Score
* ROC AUC

이러한 지표는 모델의 전반적인 성능을 빠르게 이해하는 데 도움을 준다.

---

### 2) ROC Curve 시각화

ROC Curve를 통해 다음을 확인할 수 있다.

* 모델의 분류 성능
* Threshold 변화에 따른 성능
* 모델의 구분 능력

이는 모델의 **분류 성능을 시각적으로 이해하는 데 매우 중요한 지표**이다.

---

### 3) Confusion Matrix Heatmap

Confusion Matrix를 Heatmap 형태로 제공하여 다음을 직관적으로 확인할 수 있다.

* 실제 이탈 고객을 얼마나 잘 잡았는지
* 오탐(False Positive) 발생 정도
* 미탐(False Negative) 발생 정도

---

### 4) Feature Importance 제공

Random Forest 모델의 Feature Importance를 제공하여 다음을 확인할 수 있다.

* 어떤 변수들이 이탈 예측에 가장 큰 영향을 미치는지
* 모델의 의사결정 근거

이는 모델 해석 가능성(Explainability) 측면에서 매우 중요한 기능이다.

---

## 2.2 운영 모니터링 페이지의 장점

운영 모니터링 페이지는 **실제 서비스 운영 관점의 지표**를 제공한다.

---

### 1) 비즈니스 중심 KPI 제공

다음과 같은 운영 지표를 제공한다.

* Precision
* Recall
* F1 Score
* ROC AUC
* Threshold

특히 **Recall 중심 전략**을 강조하여 이탈 고객을 놓치지 않는 운영 방식을 보여준다.

---

### 2) 현재 운영 상태 제공

다음과 같은 운영 정보를 제공한다.

* 전체 고객 수
* 이탈 예상 고객 수
* 이탈 예상 고객 비율

이는 모델이 실제 환경에서 **얼마나 많은 고객을 타겟팅하고 있는지** 보여준다.

---

### 3) 실제 적중 현황 제공

다음 정보를 통해 모델의 실제 성과를 확인할 수 있다.

* 이탈 예측 고객 중 실제 이탈 고객
* 이탈 예측 적중률
* 실제 이탈 고객 포착률

이는 모델의 **실제 운영 효율을 평가하는 데 중요한 지표**이다.

---

### 4) 이탈 예상 고객 리스트 제공

다음과 같은 고객 정보를 제공한다.

* 고객 ID
* 지역
* 상품 유형
* 이탈 확률
* 위험 등급
* 이탈 예상 이유

이를 통해 마케팅 팀은 **리텐션 캠페인 대상 고객을 직접 확인할 수 있다.**

---

### 5) 이탈 원인 설명 기능

`prediction_reason` 변수를 통해 모델이 판단한 **이탈 이유를 설명**한다.

이는 모델의 **설명 가능성(Explainability)** 측면에서 중요한 기능이다.

---

# 3. 두 페이지의 겹치는 기능

두 페이지에는 일부 기능이 중복되어 존재한다.

중복 기능은 통합 시 정리할 필요가 있다.

---

## 1) 모델 성능 지표

두 페이지 모두 다음 지표를 제공한다.

* Accuracy
* Precision
* Recall
* F1 Score
* ROC AUC

따라서 통합 페이지에서는 **하나의 KPI 영역에서 통합 제공**하는 것이 바람직하다.

---

### 2) Confusion Matrix

두 페이지 모두 Confusion Matrix를 제공한다.

차이점은 다음과 같다.

| 페이지         | 방식          |
| ----------- | ----------- |
| 모델 성능 페이지   | Heatmap 그래프 |
| 운영 모니터링 페이지 | 데이터 테이블     |

통합 페이지에서는 **Heatmap 방식만 유지하는 것이 시각적으로 더 효과적이다.**

---

## 3) 모델 평가 정보

두 페이지 모두 다음 정보를 기반으로 모델을 평가한다.

* 테스트 데이터 기반 평가
* 예측 확률 기반 평가

따라서 평가 로직은 하나로 통합하는 것이 바람직하다.

---

# 4. 추가하면 좋은 기능

통합 페이지의 활용도를 높이기 위해 다음 기능을 추가할 수 있다.

---

## 4.1 Threshold 시뮬레이션 기능

슬라이더를 사용하여 Threshold를 변경하면 다음 값이 실시간으로 변하도록 할 수 있다.

* Precision
* Recall
* 예측 고객 수

이를 통해 운영자는 **캠페인 규모를 조정할 수 있다.**

---

## 4.2 위험도 분포 시각화

고객의 이탈 확률 분포를 히스토그램으로 표현한다.

예시

```
0.0 ~ 0.2
0.2 ~ 0.4
0.4 ~ 0.6
0.6 ~ 0.8
0.8 ~ 1.0
```

이를 통해 모델이 **어떤 확률 영역에 고객을 많이 배치하는지** 확인할 수 있다.

---

## 4.3 Lift Chart

마케팅 관점에서 중요한 지표이다.

예시

* 상위 10% 고객에서 실제 이탈률
* 상위 20% 고객에서 실제 이탈률

이를 통해 **타겟 마케팅 효율을 측정**할 수 있다.

---

## 4.4 고객 필터 기능

다음 조건으로 고객을 필터링할 수 있다.

* 지역
* 상품 유형
* 위험 등급

이를 통해 마케팅 팀이 **특정 고객군만 분석할 수 있다.**

---

## 4.5 SHAP 기반 모델 설명

Feature Importance보다 더 정교한 설명 방법이다.

개별 고객에 대해 다음을 설명할 수 있다.

* 이탈 확률이 높은 이유
* 어떤 변수들이 영향을 미쳤는지

이는 모델 신뢰도를 높이는 데 도움이 된다.

---

# 5. 제거하거나 축소 가능한 정보

통합 페이지에서는 일부 정보가 불필요하거나 중복될 수 있다.

---

## 1) Classification Report 전체 출력

classification_report는 다음 정보를 포함한다.

* precision
* recall
* f1-score

하지만 동일한 정보가 KPI 카드로 이미 제공되므로
**전체 Report 출력은 제거하는 것이 좋다.**

---

## 2) Confusion Matrix 중복 표시

현재 두 페이지에서는 다음과 같이 중복 표시가 존재한다.

* Heatmap
* 데이터 테이블

통합 페이지에서는 **Heatmap만 유지하는 것이 적절하다.**

---

## 3) 과도한 설명 문장

운영 페이지에 존재하는 긴 설명 문장은 다음과 같은 문제가 있다.

* 화면 가독성 저하
* 정보 전달 효율 감소

따라서 **핵심 요약만 남기고 간결하게 정리하는 것이 바람직하다.**

---

## 4) 테스트 데이터 수

다음 정보는 운영 관점에서 중요도가 낮다.

* 테스트 데이터 수

운영 페이지에서는 제거하거나 축소할 수 있다.

---

# 6. 통합 페이지 설계 방향

통합 페이지는 다음과 같은 구조로 구성하는 것이 바람직하다.

---

## 1) KPI 영역

* Accuracy
* Precision
* Recall
* F1 Score
* ROC AUC
* Threshold

---

## 2) 모델 성능 영역

* ROC Curve
* Confusion Matrix

---

## 3) Feature Importance

* Top Feature Importance

---

## 4) 운영 상태

* 전체 고객 수
* 이탈 예상 고객 수
* 이탈 예상 비율

---

## 5) 실제 적중 현황

* 이탈 예측 고객 중 실제 이탈
* 이탈 예측 적중률
* 실제 이탈 포착률

---

## 6) 이탈 예상 고객 리스트

다음 정보를 제공한다.

* 고객 ID
* 지역
* 상품 유형
* 보험료
* 이탈 확률
* 위험 등급
* 이탈 예상 이유

---

# 7. 결론

두 개의 페이지는 각각 다음과 같은 장점을 가진다.

| 페이지         | 핵심 장점         |
| ----------- | ------------- |
| 모델 성능 페이지   | 머신러닝 모델 성능 분석 |
| 운영 모니터링 페이지 | 실제 비즈니스 운영 상태 |

두 페이지를 통합하면 다음과 같은 장점을 얻을 수 있다.

* 모델 성능과 운영 상태를 동시에 확인 가능
* 중복 기능 제거
* 사용자 경험 개선
* 비즈니스 활용도 향상

이를 통해 **데이터 사이언스 분석과 비즈니스 운영을 연결하는 통합 대시보드**를 구축할 수 있다.
